### 1. Импорт данных и отображение первых строк

Описание задачи:
У вас есть несколько файлов CSV с данными о Супербоуле, телевизионных трансляциях и артистах полувременного шоу. Ваша задача - импортировать эти данные в виде DataFrame с помощью библиотеки pandas и отобразить первые пять строк каждого DataFrame.

#### Инструкции

Импортируйте библиотеку pandas с псевдонимом pd.
Используйте функцию pd.read_csv() для загрузки данных из каждого файла CSV в соответствующий DataFrame (super_bowls, tv, halftime_musicians).
Отобразите первые пять строк каждого DataFrame с помощью функции display().

```python
# Импортировать pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Загрузить данные CSV в DataFrames
super_bowls = pd.read_csv('datasets/super_bowls.csv')
tv = pd.read_csv('datasets/tv.csv')
halftime_musicians = pd.read_csv('datasets/halftime_musicians.csv')

# Отобразить первые пять строк каждого DataFrame
display(tv.head())
display(halftime_musicians.head())
display(super_bowls.head())
```

```python
print(halftime_musicians.head())
```

```python
# Определение отсутствующих значений в DataFrame
print("Столбцы с пропущенными значениями: ", halftime_musicians.columns[halftime_musicians.isnull().any()].tolist())

# Расчет общего количества отсутствующих значений в каждом столбце
print('\nОбщее количество пропущенных значений в каждом столбце:\n', halftime_musicians.isnull().sum())

```

To find the top 5 musicians by the number of songs they performed during halftime shows, you can follow these steps:

Group the DataFrame by the musician.
Sum the number of songs performed by each musician.
Sort the musicians based on the total number of songs.
Select the top 5 musicians.

```python
# Сгруппировать по музыканту и сложить количество песен
musician_songs = halftime_musicians.groupby('musician')['num_songs'].sum()

# Отсортировать музыкантов по общему количеству песен и выбрать пять лучших
top_5_musicians = musician_songs.sort_values(ascending=False).head()
print(top_5_musicians)
```

To find the top 5 musicians by the number of appearances in the halftime_musicians DataFrame, you can follow these steps:

Group the DataFrame by the musician's name.
Count the number of appearances for each musician.
Sort the musicians based on their number of appearances in descending order.
Select the top 5 musicians.

```python
# Сгруппировать по имени музыканта и посчитать количество выступлений
musician = halftime_musicians.groupby('musician').size()

# Выбрать пять музыкантов с наибольшим количеством выступлений
musician_sort = musician.sort_values(ascending=False)

top_5 = musician_sort.head()
print(top_5)
```

To answer the question regarding changes in viewership, TV ratings, and the cost of advertising over time, you can merge the necessary DataFrames and then analyze the trends over different Super Bowl events.

Here's how you can approach this:

Merge the tv and super_bowls DataFrames based on a common column, such as the Super Bowl number or date.
Analyze the trends over time for viewership, TV ratings, and advertising cost.

```python
# Объединить DataFrames на основе общего столбца (например, номера Супербоула или даты)
merged_dataset = pd.merge(tv, super_bowls, on='super_bowl')
merged_dataset = pd.merge(merged_dataset, halftime_musicians, on='super_bowl')

# Изменить порядок дат на обратный
merged_dataset['date'] = pd.to_datetime(merged_dataset['date'])
merged_dataset = merged_dataset.sort_values(by='date', ascending=True)

# Анализ трендов со временем для просмотров, рейтинга ТВ и стоимости рекламы


# Просмотры со временем
plt.figure(figsize=(12, 6))
plt.plot(merged_dataset['date'], merged_dataset['avg_us_viewers'], marker='o', linestyle='-', color='skyblue')
plt.xlabel('Дата')
plt.ylabel('Средние просмотры')
plt.title('Просмотры со временем')
plt.xticks(rotation=45)
plt.gca().invert_xaxis()  
plt.show()

# Рейтинг ТВ со временем
plt.figure(figsize=(12, 6))
plt.plot(merged_dataset['date'], merged_dataset['rating_household'], marker='o', linestyle='-', color='orange')
plt.xlabel('Дата')
plt.ylabel('Рейтинг ТВ (домохозяйства)')
plt.title('Рейтинг ТВ со временем')
plt.xticks(rotation=45)
plt.gca().invert_xaxis()  
plt.show()

# Стоимость рекламы со временем
plt.figure(figsize=(12, 6))
plt.plot(merged_dataset['date'], merged_dataset['ad_cost'], marker='o', linestyle='-', color='green')
plt.xlabel('Дата')
plt.ylabel('Стоимость рекламы')
plt.title('Стоимость рекламы со временем')
plt.xticks(rotation=45)
plt.gca().invert_xaxis()  
plt.show()
```

To determine the most extreme outcomes of the games, you might want to consider various metrics such as the largest winning point difference, the highest combined points scored, or the lowest viewership.

Here's how you can find the most extreme outcomes using these metrics:

Largest winning point difference: Identify the game(s) with the largest difference in winning points.
Highest combined points scored: Identify the game(s) with the highest total points scored by both teams.
Lowest viewership: Identify the game(s) with the lowest viewership.


```python
merged_sp_tv = pd.merge(super_bowls, tv, on='super_bowl')

# Самая большая разница в очках между победителем и проигравшим
max_point_diff = merged_sp_tv.loc[merged_sp_tv['difference_pts'].idxmax()]
game_max_diff = max_point_diff['super_bowl']
point_difference = max_point_diff['difference_pts']

# Самое высокое количество очков, набранных в совокупности
max_combined_pts = merged_sp_tv.loc[merged_sp_tv['combined_pts'].idxmax()]
game_max_combined_pts = max_combined_pts['super_bowl']
combined_points = max_combined_pts['combined_pts']

# Самый низкий уровень просмотров
min_avg_viewers = merged_sp_tv.loc[merged_sp_tv['avg_us_viewers'].idxmin()]
game_min_viewers = min_avg_viewers['super_bowl']
average_viewers = min_avg_viewers['avg_us_viewers']

print("Наиболее экстремальные результаты:")
print(f"Самая большая разница в очках между победителем и проигравшим:")
print(f"Игра: {game_max_diff}, Разница: {point_difference}\n")

print(f"Самое высокое количество очков, набранных в совокупности:")
print(f"Игра: {game_max_combined_pts}, Комбинированное количество очков: {combined_points}\n")

print(f"Самый низкий уровень просмотров:")
print(f"Игра: {game_min_viewers}, Среднее количество зрителей: {average_viewers}\n")
```
